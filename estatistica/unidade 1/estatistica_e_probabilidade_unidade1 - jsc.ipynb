**Data Dictionary**

| Variável | Definição                                 | Chave                                            | Notas                                                                                                                                       |
|----------|--------------------------------------------|------------------------------------------------|---------------------------------------------------------------------------------------------------------------------------------------------|
| CD_DISCENTE | Identificador do estudante| Numérica| - |
|ANO_INGRESSO |Ano de ingresso do discente|Numérica|Não há valor diferente de 2023 pela natureza do dataset|
| SEMESTRE_INGRESSO | Semestre onde o discente dará início às aulas|Numérica| - |
|TIPO_INGRESSO| Meio de ingresso |Texto| Refere-se à forma que o discente ingressou, seja SISU, transferência, etc |
|CAMPUS|Campus onde o discente está alocado|Texto|A UFPE proporciona 3 opções: Reitor Joaquim Amazonas (campus Recife), Agreste, e Vitória
|COD_CURSO |Identificador do curso |Numérica| - |
|CURSO |Nome do curso|Texto |Cursos do Agreste e de Vitória recebem as siglas CAA e CAV respectivamente|
|NATURALIDADE|Cidade onde o estudante nasceu|Texto| - |
|UF_NATURALIDADE|Unidade federativa onde o estudante nasceu|Texto|-|
|PAIS_NATURALIDADE|País onde o estudante nasceu|Texto|-|
|BAIRRO_ENDERECO|Bairro onde o estudante mora no momento da matrícula|Texto|-|
|CIDADE_ENDERECO|Cidade onde o estudante mora no momento da matrícula|Texto|-|
|UF_ENDERECO| Estado onde o estudante mora no momento da matrícula|Texto|-|
|SEXO |Sexo do discente|Caractere|-|
|ANO_NASCIMENTO |Ano de nascimento do discente|Numérica|Existe no dataset uma ocorrência do valor 2023, acredito que a intenção era preencher com 2003; fora esse caso, o maior valor é 2006|
|COTA |Se o discente fez ou não o uso de cotas no processo de ingresso|Caractere|-|
|TIPO_COTA| Tipo da cota usada |Texto| Discentes que não usaram cotas constam com o valor A0, que indica Ampla Concorrência |
|ESCOLARIDADE_PAI| Nível de escolaridade do pai do discente |Texto| Existem valores nulos nessa coluna |
|ESCOLARIDADE_MAE| Nível de escolaridade da mãe do discente |Texto| Existem valores nulos nessa coluna |
|DEFICIENCIA |Se o discente é PCD ou não|Caractere|-|

## **Probabilidade simples**



```python
pMulher = len(discentes[discentes['SEXO']== 'F']) / len(discentes)
print('A probabilidade de ser uma ingressante mulher é igual a: '+ str(round(pMulher*100,4)) + '%')
```

```python
pPCD = len(discentes[discentes['DEFICIENCIA']== 'S']) / len(discentes)
print('A probabilidade do ingressante ser PCD é igual a: '+ str(round(pPCD*100,4)) + '%')
```

```python
pCota = len(discentes[discentes['COTA']== 'S']) / len(discentes)
print('A probabilidade do ingressante ser cotista é igual a: '+ str(round(pCota*100,4)) + '%')
```

```python
cotaRacial = len(discentes[discentes['TIPO_COTA'] == 'L2']) + len(discentes[discentes['TIPO_COTA'] == 'L4']) + len(discentes[discentes['TIPO_COTA'] == 'L6']) + len(discentes[discentes['TIPO_COTA'] == 'L7']) + len(discentes[discentes['TIPO_COTA'] == 'L8']) + len(discentes[discentes['TIPO_COTA'] == 'L10']) + len(discentes[discentes['TIPO_COTA'] == 'L11']) + len(discentes[discentes['TIPO_COTA'] == 'L14'])
#soma todos os ingressantes que utilizam de cotas direcionadas a pretos, pardos, indígenas e quilombolas, l2, l4, l6, l7, l8, l10, l11, l14

pCotaRacial = cotaRacial / len(discentes)
print('A probabilidade de um ingressante usar cota racial é igual a: ' + str(round(pCotaRacial*100,4)) + '%')
```

```python
fromRec = discentes.NATURALIDADE.value_counts()['RECIFE']
outRec = len(discentes) - fromRec

pOutRec = outRec / len(discentes)
print('A probabilidade de um ingressante não ser natural de Recife é igual a: ' + str(round(pOutRec*100,4)) + '%')
```

```python
campusRec = discentes.CAMPUS.value_counts()['REITOR JOAQUIM AMAZONAS']

pCampusRec = campusRec / len(discentes)
print('A probabilidade de um ingressante estudar em Recife é igual a: ' + str(round(pCampusRec*100,4)) + '%')
```

```python
entrarSisu = discentes.TIPO_INGRESSO.value_counts()['SISU']

pEntrarSisu = entrarSisu / len(discentes)
print('A probabilidade de um estudante ingressar pelo SISU é igual a: ' + str(round(pEntrarSisu*100,4)) + '%')
```



---


## **Probabilidade com união**

```python
campusDesc = len(discentes[(discentes['CAMPUS'] == 'AGRESTE') | (discentes['CAMPUS'] == 'VITORIA')])

pCampusDesc = round(campusDesc / len(discentes), 4)
print('A probabilidade do estudante escolher um campus descentralizado é de: '+ str(round(pCampusDesc*100,4)) + '%')
```

```python
paiSup = len(discentes[(discentes['ESCOLARIDADE_PAI'] == 'ENSINO SUPERIOR') | (discentes['ESCOLARIDADE_PAI'] == 'SUPERIOR INCOMPLETO')])
maeSup = len(discentes[(discentes['ESCOLARIDADE_MAE'] == 'ENSINO SUPERIOR') | (discentes['ESCOLARIDADE_MAE'] == 'SUPERIOR INCOMPLETO')])

paisSup = paiSup | maeSup

pPaisSup = round(paisSup / len(discentes), 4)
print('A probabilidade do estudante ter pelo menos um dos pais que alcançou o nível superior é de: '+ str(round(pPaisSup*100,4)) + '%')
```

```python
paisSupComp = len(discentes[(discentes['ESCOLARIDADE_PAI'] == 'ENSINO SUPERIOR') | (discentes['ESCOLARIDADE_MAE'] == 'ENSINO SUPERIOR')])

pPaisSupComp = round(paisSupComp / len(discentes), 4)
print('A probabilidade do estudante ter pelo menos um dos pais com ensino superior completo é de: '+ str(round(pPaisSupComp*100,4)) + '%')
```

```python
cursoCIn = len(discentes[(discentes['CURSO'] == 'CIÊNCIA DA COMPUTAÇÃO - CIN') | (discentes['CURSO'] == 'ENGENHARIA DA COMPUTAÇÃO') | (discentes['CURSO'] == 'SISTEMAS DE INFORMAÇÃO')])

pCursoCIn = round(cursoCIn / campusRec, 4)
#o uso específico do campus de Recife é porque cursos "oficialmente" de tecnologia estão presentes apenas no cin
print('A probabilidade do estudante ingressar no campus de Recife num curso da área de informática é de: '+ str(round(pCursoCIn*100,4)) + '%')
```



---


##**Probabilidade com intersecção**

```python
pcdMed = len(discentes[(discentes['DEFICIENCIA'] == 'S') & ((discentes['CURSO'] =='MEDICINA') | (discentes['CURSO'] == 'MEDICINA - CAA'))])

pPCDMed = pcdMed / len(discentes)
print('A probabilidade de ser um estudante de Medicina ter uma deficiência é de: '+ str(round(pPCDMed*100, 4)) + '%')
```

```python
cotaRacialMed = len(discentes[(((discentes['CURSO'] =='MEDICINA') | (discentes['CURSO'] == 'MEDICINA - CAA'))) & cotaRacial])
pCotaRMed = cotaRacialMed / len(discentes)

print('A probabilidade de um estudante usar cota racial e estudar Medicina é de: '+ str(round(pCotaRMed*100, 4)) + '%')
```

```python
pcdAgr = len(discentes[(discentes['DEFICIENCIA'] == 'S') & (discentes['CAMPUS'] =='AGRESTE')])
pPCDAgr = pcdAgr / len(discentes)

print('A probabilidade de um estudante PCD e estudar no Campus Agreste é de: '+ str(round(pPCDAgr*100, 4)) + '%')
```



---


##**Probabilidade com complemento**



```python
paisSemSup = len(discentes) - paisSup

pPaisSemSup = paisSemSup / len(discentes)
print('A probabilidade de um estudante ser da primeira geração de sua família ao alcançar o ensino superior é de: '+ str(round(pPaisSemSup*100, 4)) + '%')
```

```python
pMuPaisSSup = pMulher - pPaisSup

print('A probabilidade de um estudante ser mulher e ter pais sem Ensino Superior é de: '+ str(round(pMuPaisSSup*100, 4)) + '%')
```

```python
pNaoEntrarSisu = (len(discentes) - entrarSisu) / len(discentes)

print('A probabilidade de um estudante não ingressar pelo SISU é igual a: ' + str(round(pNaoEntrarSisu*100,4)) + '%')
```



---


## **Probabilidade da diferença**



```python
paiSupMaeNao = len(discentes[(discentes['ESCOLARIDADE_PAI'] == 'ENSINO SUPERIOR') & (discentes['ESCOLARIDADE_MAE'] != 'ENSINO SUPERIOR')])
pPaiSupMaeNao = paiSupMaeNao / len(discentes)

print('A probabilidade do pai ter ensino superior e a mãe não é: ' + str(round((pPaiSupMaeNao*100),2))+'%')
```

```python
maeSupPaiNao = len(discentes[(discentes['ESCOLARIDADE_MAE'] == 'ENSINO SUPERIOR') & (discentes['ESCOLARIDADE_PAI'] != 'ENSINO SUPERIOR')])
pMaeSupPaiNao = maeSupPaiNao / len(discentes)

print('A probabilidade do pai não ter ensino superior e a mãe ter é: ' + str(round((pMaeSupPaiNao*100),2))+'%')
```



---


## **Probabilidade condicional**

```python
pCondCota = pCotaRacial / pCota
print('A probabilidade do estudante usar cotas raciais sabendo que é cotista é: ' + str(round((pCondCota*100),2))+'%')
```

```python
pCondPCDMed = pPCDMed / pPCD
print('A probabilidade do estudante ser PCD sabendo que é estudante de Medicina é: ' + str(round((pCondPCDMed*100),2))+'%')
```

```python
pCondCIn = pCursoCIn / pCampusRec
print('A probabilidade do ingressante estudar no CIn sabendo que é do campus Recife  é: ' + str(round((pCondCIn*100),2))+'%')
```



---


## **Probabilidade com teorema de Bayes**

```python
pbPaisSupComp = (pPaisSupComp * pPaisSup) / len(discentes)
print('A probabilidade de ter pais com ensino superior tal que pelo menos um de seus pais tenha alcançado o nível superior é: '+ str(round(pbPaisSupComp*100, 4)) + '%')
```

```python
pbMaeSupPaiNao = (pMaeSupPaiNao * pPaisSupComp) / pPaisSup
print('A probabilidade da mãe ensino superior tal que pelo menos um de seus pais tenha terminado o nível superior é: '+ str(round(pbMaeSupPaiNao*100, 4)) + '%')
```

```python
pbPaiSupMaeNao = (pPaiSupMaeNao * pPaisSupComp) / pPaisSup
print('A probabilidade do pai ensino superior tal que pelo menos um de seus pais tenha terminado o nível superior é: '+ str(round(pbPaiSupMaeNao*100, 4)) + '%')
```
